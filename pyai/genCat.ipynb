```python
def better_split_subcategory_name(subcategory: str) -> str:
    """

    NOTE: Only called if file doesn't exist.
    Split concatenated words in a subcategory name into readable words using a lookup table.
    Args:
        subcategory (str): The original subcategory name.
    Returns:
        str: The cleaned subcategory name.
    """
    # Check in the lookup table first
    if subcategory in SUBCATEGORY_LOOKUP:
        return SUBCATEGORY_LOOKUP[subcategory]
    
    # If not in lookup, apply regex-based splitting
    subcategory = re.sub(r'([a-z])([A-Z])', r'\1 \2', subcategory)  # camelCase
    subcategory = re.sub(r'([A-Z])([A-Z][a-z])', r'\1 \2', subcategory)  # Acronyms + word
    subcategory = re.sub(r'([a-zA-Z])([0-9])', r'\1 \2', subcategory)  # Letter + digit
    subcategory = re.sub(r'([0-9])([a-zA-Z])', r'\1 \2', subcategory)  # Digit + letter
    subcategory = re.sub(r'[_\-]', ' ', subcategory)  # Replace underscores and hyphens
    return subcategory.strip().title()



def extract_categories_and_subcategories(news_data: pd.DataFrame) -> dict:
    """
    Extract and clean categories and subcategories from a News DataFrame.
    Args:
        news_data (pd.DataFrame): DataFrame containing 'category' and 'subcategory' columns.
    Returns:
        dict: A dictionary where keys are categories and values are cleaned lists of subcategories.
    """
    print("Preprocessing data...")
    cleaned_data = news_data.dropna(subset=['category', 'subcategory']).drop_duplicates()

    print("Grouping data by category...")
    grouped_data = cleaned_data.groupby('category')['subcategory']

    print("Processing groups to extract and clean subcategories...")
    result = {}
    for category, subcategories in grouped_data:
        try:
            unique_subcategories = sorted(subcategories.dropna().unique())
            cleaned_subcategories = [better_split_subcategory_name(subcat) for subcat in unique_subcategories]
            result[category] = cleaned_subcategories
            print(f"Processed '{category}': {cleaned_subcategories[:5]}...")  # Debug: First 5 cleaned subcategories
        except Exception as e:
            print(f"Error processing category '{category}': {e}")
            raise e

    print("Processing complete!")
    return result

def find_missing_subcategories(original_subcategories: list, lookup_table: dict) -> list:
    """
    Find subcategories (original keys) not present in the lookup table.

    Args:
        original_subcategories (list): List of all original subcategories to check.
        lookup_table (dict): The existing lookup table mapping original subcategories to cleaned names.

    Returns:
        list: Original subcategories that are not in the lookup table.
    """
    missing_subcategories = [subcat for subcat in original_subcategories if subcat not in lookup_table.keys()]
    return missing_subcategories

```
